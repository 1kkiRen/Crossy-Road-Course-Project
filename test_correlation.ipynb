## Using Correlation to construct a bounding box and collect info about object location

```python
screenshot = cv2.imread('screen1.png', cv2.IMREAD_COLOR)
screenshot_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGR2GRAY)

threshold = 0.86

detected_boxes = []

for template_file in template_files:

    template = cv2.imread(template_file, cv2.IMREAD_COLOR)
    template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

    template_height, template_width = template_gray.shape

    # Template matching using Normalized Cross-Correlation (NCC)
    result = cv2.matchTemplate(
        screenshot_gray, template_gray, cv2.TM_CCOEFF_NORMED)
    locations = np.where(result >= threshold)

    # Iterate over detected locations
    for pt in zip(*locations[::-1]):
        top_left = pt
        bottom_right = (top_left[0] + template_width,
                        top_left[1] + template_height)

        cv2.rectangle(screenshot, top_left, bottom_right,
                      (0, 255, 0), 2)
        detected_boxes.append((template_file, top_left, bottom_right))

        print(f"Detected object from {template_file}")
        print(f" at {top_left} with bounding box from")
        print(f"{top_left} to {bottom_right}")

cv2.imshow('Detected Objects', screenshot)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('detected_objects.jpg', screenshot)
```
